```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


%matplotlib inline
```

```python
df = pd.read_csv("tv.csv") #memanggil data set csv
df.head() #menampilkan 5 data awal
```

```python
df.info() #menampilkan informasi 
```

```python
df.shape #menghitung baris dan kolom
duplicate_row_df = df[df.duplicated()] ## untuk melihat jumlah baris yang memiliki duplikat dengan diwakili oleh variabel duplicate_rows_df. Pada saat data duplikat ini akan dihapus, maka variabel tersebutlah yang akan dihapus.
df.count() #menghitung atribut
```

```python
df.shape
```

```python
print("mean")
print(df.mean(numeric_only=True)) #mengambil nilai rata rata

print("\nMedian:")
print(df.median(numeric_only=True)) #mengambil nilai tengah dari data csv

print("\nStandard Deviation: ")
print(df.std(numeric_only=True))

print("\nVariance:")
print(df.var(numeric_only=True)) #hanya menghitung kolom yang bertipe angka (int64 atau float64)

print("\nMinimum:")
print(df.min(numeric_only=True)) #mengambil nilai minumum

print("\nMaximum:")
print(df.max(numeric_only=True)) #mengambil nilai maximum
```

```python
df.describe() #untuk mengambil data yag sudah diolah sebelumnya
```

```python
df.plot.scatter(x='av_rating', y='seasonNumber', alpha=0.5) #Perintah untuk membuat grafik sebar (scatter plot). Grafik ini digunakan untuk melihat hubungan antara dua variabel numerik.
                                                 #nilai (x) Meletakkan nilai rating rata-rata pada sumbu horizontal (mendatar).
                                                 #nilai (y) Meletakkan jumlah musim pada sumbu vertikal (tegak).
plt.title('Scatter Plot Average Rating vs Number of Seasons') #Memberikan judul dan label keterangan pada grafik agar mudah dibaca.
plt.xlabel('Average Rating') 
plt.ylabel('Number of Seasons')
plt.show()
```

```python
fig = px.scatter(df, 
                 x='av_rating', 
                 y='seasonNumber', 
                 hover_name='title',  
                 color='genres',      
                 title='Interaktif: Rating vs Jumlah Musim')

fig.show()
```
